```python
# Memasukkan Inputan Data
crate = float(input('Masukkan crate artifakmu traveller: '))
cdmg = float(input('Masukkan cdmg artifakmu traveller: '))

# Menghitung Crit Value
cv= (crate*2) + cdmg

# Menampilkan Hasil Crit Value
print('Crit value artifakmu sebesar:', cv, '\n')

if cv>=60:
  print('Lain kali kalo mau ngegarem otaknya diservis dulu ya, jangan kebanyakan halu!')
elif (40<=cv<=55):
  print('Hati-hati kalo keluar rumah, takutnya ditabrak truk masuk isekai!')
elif (20<=cv<=39):
  print('Artifakmu sudah bagus,Traveller!')
elif (10<=cv<=19):
  print('Sudah lumayan bagus, bisa dikeep dulu, nanti kalau ada yang lebih bagus baru ditumbalin!')
else:
  print('Lihat dulu stat lainnya ya, kalau masih bagus jangan ditumbal dulu!')
```
